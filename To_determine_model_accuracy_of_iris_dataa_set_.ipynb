```python

iris = love    # as sklearn work well in numpy we again need to convert in to numpy 
```

```python
x = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=1)
```

as we took test size as 33% so out of total data of 150

```python
print(x_train.shape)   
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
classifier_knn = KNeighborsClassifier(n_neighbors = 7)
classifier_knn.fit(x_train,y_train)
y_pred = classifier_knn.predict(x_test)
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

***model accuracy for different valuse of k from 2 to 20***

```python
K = []
training = []
test = []
scores = {}
  
for k in range(2, 21):
    clf = KNeighborsClassifier(n_neighbors = k)
    clf.fit(x_train, y_train)
  
    training_score = clf.score(x_train, y_train)
    test_score = clf.score(x_test, y_test)
    K.append(k)
  
    training.append(training_score)
    test.append(test_score)
    scores[k] = [training_score, test_score]
```

```python
for keys, values in scores.items():
    print(keys, ':', values)
```
